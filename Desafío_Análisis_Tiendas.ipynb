```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload dataframes to ensure they are correctly initialized
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

facturacion_tiendas = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

## Convertir el diccionaro en un DataFrame de pandas
df_facturacion = pd.DataFrame(list(facturacion_tiendas.items()), columns=["Tienda", "Facturación"])

##Generar el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x="Tienda", y="Facturación", data=df_facturacion)
plt.title("Facturación por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Facturación")
plt.tight_layout()
plt.show()

## El código original para encontrar la tienda con mayor facturacion se mantien igual
mayor_Fact = 0
tienda_mayor_fact = ""

# Renaming loop variable to avoid conflict with DataFrame names
for tienda_name, facturacion in facturacion_tiendas.items():
  if facturacion > mayor_Fact:
    mayor_Fact = facturacion
    tienda_mayor_fact = tienda_name

print(f"\n La tienda con la mayor facturacion es: {tienda_mayor_fact} con un valor total de: ${mayor_Fact}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer cada tienda y agregar una columna para identificar la tienda
tienda = pd.read_csv(url)
tienda["Tienda"] = "Tienda 1"

tienda2 = pd.read_csv(url2)
tienda2['Tienda'] = 'Tienda 2'


tienda3 = pd.read_csv(url3)
tienda3['Tienda'] = 'Tienda 3'

tienda4 = pd.read_csv(url4)
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los DataFrames
ventas_totales = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

ventas_totales.head()

```

```python
ventas_por_tienda_producto = ventas_totales.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad de Ventas')

max_ventas = ventas_por_tienda_producto.groupby('Tienda')['Cantidad de Ventas'].idxmax()
mas_vendido_por_tienda = ventas_por_tienda_producto.loc[max_ventas]

min_ventas = ventas_por_tienda_producto.groupby('Tienda')['Cantidad de Ventas'].idxmin()
menos_vendido_por_tienda = ventas_por_tienda_producto.loc[min_ventas]

print("\nProducto mas Vendido por Tienda:")
print(mas_vendido_por_tienda)

print("\nProducto menos Vendido por Tienda:")
print(menos_vendido_por_tienda)
```

```python
# Get top 2 products per store
top_2_ventas = ventas_por_tienda_producto.groupby('Tienda').apply(lambda x: x.nlargest(2, 'Cantidad de Ventas')).reset_index(drop=True)

# Get bottom 2 products per store
bottom_2_ventas = ventas_por_tienda_producto.groupby('Tienda').apply(lambda x: x.nsmallest(2, 'Cantidad de Ventas')).reset_index(drop=True)

# Combine the top 2 and bottom 2 into a single DataFrame
combined_ventas = pd.concat([top_2_ventas, bottom_2_ventas])

print("Combined data for top 2 and bottom 2 products per store:")
display(combined_ventas.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.barplot(x='Tienda', y='Cantidad de Ventas', hue='Producto', data=combined_ventas, palette='tab10')
plt.title('Top 2 and Bottom 2 Sold Products per Store')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Ventas')
plt.legend(title='Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 1. Distribución General de Ventas (Scatter Plot):

El gráfico de dispersión global que muestra todas las ventas por latitud y longitud, diferenciadas por tienda, revela una concentración significativa de puntos en ciertas áreas geográficas. Esto sugiere que las ventas no están distribuidas uniformemente y que existen *clusters* de actividad comercial. La superposición de colores indica que las tiendas a menudo comparten áreas geográficas de operación, lo que podría implicar competencia directa en esos puntos o una estrategia de saturación de mercado en zonas de alta demanda.

- **Patrones de Ventas por Ubicación**: Se observa que la mayoría de las ventas se concentran en regiones específicas. Esto puede ser debido a la densidad poblacional, accesibilidad o poder adquisitivo en esas áreas. Las ventas no cubren toda la extensión geográfica de Colombia, sino que se agrupan en las principales ciudades o regiones metropolitanas.

- **Superposición de Tiendas**: Hay una clara superposición de las ventas de diferentes tiendas en las mismas ubicaciones geográficas. Por ejemplo, en las áreas de mayor concentración, es común ver puntos de venta de 'Tienda 1', 'Tienda 2', 'Tienda 3' y 'Tienda 4' mezclados. Esto sugiere que no hay una especialización geográfica estricta por tienda, sino que todas compiten por los mismos mercados principales.

Los mapas de calor (KDE plots) proporcionan una visión más clara de las áreas de alta concentración de ventas.

La ubicación geográfica claramente influye en el rendimiento general de las tiendas, aunque la granularidad de los datos ('lat' y 'lon' de la compra, no de la tienda física) hace que la interpretación sea sobre el lugar donde ocurren las ventas, no directamente sobre la ubicación de la tienda que las realiza.

- **Ingresos**: Las tiendas con mayor facturación (Tienda 1, según el análisis de facturación previo) son las que logran una mayor concentración de ventas en los puntos geográficos más activos.
