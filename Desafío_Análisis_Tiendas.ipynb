### **1. Análisis de Facturación**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

suma = tienda['Precio'].sum()
suma2 = tienda2['Precio'].sum()
suma3 = tienda3['Precio'].sum()
suma4 = tienda4['Precio'].sum()
```

```python
facturacion_tiendas = {
    'tienda_1': suma,
    'tienda 2': suma2,
    'tienda 3': suma3,
    'tienda 4': suma4,

    }

mayor_Fact = 0
tienda_mayor_fact = ""

for tienda, facturacion in facturacion_tiendas.items():
  if facturacion > mayor_Fact:
    mayor_Fact = facturacion
    tienda_mayor_fact = tienda


print(f"La tienda con la mayor facturacion es: {tienda_mayor_fact} con una facturacion de: {mayor_Fact}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload dataframes to ensure they are correctly initialized
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

facturacion_tiendas = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

## Convertir el diccionaro en un DataFrame de pandas
df_facturacion = pd.DataFrame(list(facturacion_tiendas.items()), columns=["Tienda", "Facturación"])

##Generar el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x="Tienda", y="Facturación", data=df_facturacion)
plt.title("Facturación por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Facturación")
plt.tight_layout()
plt.show()

## El código original para encontrar la tienda con mayor facturacion se mantien igual
mayor_Fact = 0
tienda_mayor_fact = ""

# Renaming loop variable to avoid conflict with DataFrame names
for tienda_name, facturacion in facturacion_tiendas.items():
  if facturacion > mayor_Fact:
    mayor_Fact = facturacion
    tienda_mayor_fact = tienda_name

print(f"\n La tienda con la mayor facturacion es: {tienda_mayor_fact} con un valor total de: ${mayor_Fact}")
```

##**2. Ventas por Categoría**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer cada tienda y agregar una columna para identificar la tienda
tienda = pd.read_csv(url)
tienda["Tienda"] = "Tienda 1" # Changed to a single 'Tienda' column

tienda2 = pd.read_csv(url2)
tienda2["Tienda"] = "Tienda 2" # Changed to a single 'Tienda' column

tienda3 = pd.read_csv(url3)
tienda3["Tienda"] = "Tienda 3" # Changed to a single 'Tienda' column

tienda4 = pd.read_csv(url4)
tienda4["Tienda"] = "Tienda 4" # Changed to a single 'Tienda' column

# Unir todos los DataFrames
ventas_totales = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Agrupar por tienda y categoria el producto, contando las ventas
conteo_por_tienda_categoria = ventas_totales.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad de Ventas')

# Encontrar la categoría más popular por tienda
categoria_mas_popular = conteo_por_tienda_categoria.loc[conteo_por_tienda_categoria.groupby('Tienda')['Cantidad de Ventas'].idxmax()]

print("Cantidad de productos vendidos por categoría en cada tienda:")
print(conteo_por_tienda_categoria)

print("\nCategoría más popular por tienda:")
print(categoria_mas_popular)
```

##**3. Calificacion promedio de la tienda**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer cada tienda y agregar una columna para identificar la tienda
tienda = pd.read_csv(url)
tienda["Tienda"] = "Tienda 1"

tienda2 = pd.read_csv(url2)
tienda2['Tienda'] = 'Tienda 2'


tienda3 = pd.read_csv(url3)
tienda3['Tienda'] = 'Tienda 3'

tienda4 = pd.read_csv(url4)
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los DataFrames
ventas_totales = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Calcular la calificacion promedio por tienda
calificacion_promedio_por_tienda = ventas_totales.groupby('Tienda')['Calificación'].mean().reset_index(name='Calificación promedio')

print("Calificación promedio por tienda:")
print(calificacion_promedio_por_tienda)
```
